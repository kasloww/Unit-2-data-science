Result in Application Fee for Boroughs

```python
import matplotlib.pyplot as plt
import pandas as pd

# opens csv file
df = pd.read_csv('dobjobapps.csv')
```

```python
df.head()
df['Initial Cost'] = df['Initial Cost'].str.strip('$').astype(float)
```

```python

data1 = df.groupby(['Borough', 'Fee Status'])['Borough'].count()


data1.plot(kind='bar')
```

```python
df['GIS_NTA_NAME'].unique().__len__()
```

Avg. Initial Cost in Brooklyn

```python
# Filters only Brooklyn Areas
data2 = df.query('Borough == "BROOKLYN"').groupby(['GIS_NTA_NAME'])['Initial Cost'].mean()

data2.plot(kind='bar')
```
