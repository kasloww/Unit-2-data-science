```python
# bar graph
import numpy as np

plt.style.use('_mpl-gallery')

x = 1 + np.arange(8)
y = [2, 3, 4, 5, 6, 7, 8, 9]

fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", color="purple", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()

# pi chart
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "color":"white", "edgecolor": "purple"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
